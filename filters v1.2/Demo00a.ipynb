```python
hdmi_in = HDMI('in')
hdmi_in.start()
hdmi_out = HDMI('out', frame_list=hdmi_in.frame_list)
hdmi_out.mode(3)#2 is for camera
hdmi_out.start()
```

```python
R0 =Register(0)
R1 =Register(1)
R2 =Register(2)
R3 =Register(3)
R4 =Register(4)
R5 =Register(5)
R6 =Register(6)
R7 =Register(7)
import ipywidgets as widgets


R0_s = widgets.IntSlider(
    value=255,
    min=0,
    max=255,
    step=1,
    description='R0:',
    disabled=False,
    continuous_update=True,
    orientation='vertical',
    readout=True,
    readout_format='i',
    slider_color='red'
)
R1_s = widgets.IntSlider(
    value=255,
    min=0,
    max=255,
    step=1,
    description='R1:',
    disabled=False,
    continuous_update=True,
    orientation='vertical',
    readout=True,
    readout_format='i',
    slider_color='green'
)
R2_s = widgets.IntSlider(
    value=255,
    min=0,
    max=255,
    step=1,
    description='R2:',
    disabled=False,
    continuous_update=True,
    orientation='vertical',
    readout=True,
    readout_format='i',
    slider_color='blue'
)
R3_s = widgets.IntSlider(
    value=255,
    min=0,
    max=255,
    step=1,
    description='R3:',
    disabled=False,
    continuous_update=True,
    orientation='vertical',
    readout=True,
    readout_format='i',
    slider_color='yellow'
)

R4_s = widgets.IntSlider(
    value=0,
    min=0,
    max=4,
    step=1,
    description='R4:',
    disabled=False,
    continuous_update=True,
    orientation='vertical',
    readout=True,
    readout_format='i',
    slider_color='yellow'
)

R5_s = widgets.IntSlider(
    value=255,
    min=0,
    max=255,
    step=1,
    description='R5:',
    disabled=False,
    continuous_update=True,
    orientation='vertical',
    readout=True,
    readout_format='i',
    slider_color='yellow'
)

R6_s = widgets.IntSlider(
    value=255,
    min=0,
    max=255,
    step=1,
    description='R6:',
    disabled=False,
    continuous_update=True,
    orientation='vertical',
    readout=True,
    readout_format='i',
    slider_color='yellow'
)

R7_s = widgets.IntSlider(
    value=255,
    min=0,
    max=255,
    step=1,
    description='R7:',
    disabled=False,
    continuous_update=True,
    orientation='vertical',
    readout=True,
    readout_format='i',
    slider_color='yellow'
)

def update_r0(*args):
    R0.write(R0_s.value)
R0_s.observe(update_r0, 'value')
def update_r1(*args):
    R1.write(R1_s.value)
R1_s.observe(update_r1, 'value')
def update_r2(*args):
    R2.write(R2_s.value)
R2_s.observe(update_r2, 'value')
def update_r3(*args):
    R3.write(R3_s.value)
R3_s.observe(update_r3, 'value')
def update_r4(*args):
    R4.write(R4_s.value)
R4_s.observe(update_r4, 'value')
def update_r5(*args):
    R5.write(R5_s.value)
R5_s.observe(update_r5, 'value')
def update_r6(*args):
    R6.write(R6_s.value)
R6_s.observe(update_r6, 'value')
def update_r7(*args):
    R7.write(R7_s.value)
R7_s.observe(update_r7, 'value')
```

```python
from IPython.display import clear_output
from ipywidgets import Button, HBox, VBox

words = ['HDMI Reset','Pass', 'RGB', 'Mirror', 'Box Blur','Motion Blur', 'Text Scroll']
items = [Button(description=w) for w in words]


def on_hdmi_clicked(b):
    hdmi_out.stop()
    hdmi_in.stop()
    hdmi_out.start()
    hdmi_in.start()
def on_pass_clicked(b):
    Bitstream_Part("pass_p.bit").download()
def on_rgb_clicked(b):
    Bitstream_Part("rgb_p.bit").download()
    R0.write(255)
    R1.write(255)
    R2.write(255)
    R0_s.value = 255
    R0_s.max = 255
    R0_s.description='Red:'
    R1_s.value = 255
    R1_s.max = 255
    R1_s.description='Green:'
    R2_s.value = 255
    R2_s.max = 255
    R2_s.description='Blue:'
    R3_s.disabled = True
    R3_s.description='Disabled'
def on_Mirror_clicked(b):
    Bitstream_Part("mirror_p.bit").download()
def on_Box_Blur_clicked(b):
    Bitstream_Part("blur3x3_p.bit").download()
def on_Motion_clicked(b):
    Bitstream_Part("motion_p.bit").download()
    R0.write(0)
    R0_s.max = 512
    R0_s.value = 0
    R0_s.description='blur length:'
def on_Scroll_clicked(b):
    Bitstream_Part("text_scroll_p.bit").download()
    R0.write(0)
    R0_s.max = (1000)
    R0_s.value = 0
    R0_s.description='Y axis'
    R1.write(150)
    R1_s.max = (400)
    R1_s.min = (130)
    R1_s.value = 150
    R1_s.description='Left Limit'
    R2.write(1150)
    R2_s.max = 1200
    R2_s.min = (500)
    R2_s.value = 1150
    R2_s.description='Right Limit' 
    R3.write(32)
    R3_s.max = (32)
    R3_s.min = (0)
    R3_s.value = 32
    R3_s.description='# Chars'
    R4_s.description='Text size'
    R5_s.description='Red'
    R6_s.description='Blue'
    R7_s.description='Green'
    
    R3_s.disabled = False

items[0].on_click(on_hdmi_clicked)
items[1].on_click(on_pass_clicked)
items[2].on_click(on_rgb_clicked)
items[3].on_click(on_Mirror_clicked)
items[4].on_click(on_Box_Blur_clicked)
items[5].on_click(on_Motion_clicked)
items[6].on_click(on_Scroll_clicked)
```
